Build a regression model.

```python
#Import packages
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
#Import data from previous CSV
bikesLA_data = pd.read_csv('data/final_data.csv')
bikesLA_data
```

```python
#Show basic information from dataframe
bikesLA_data.info()
```

```python
#Decided to change price information to add value to the model.
bikesLA_data['price']= np.where(bikesLA_data['price'] == '$',0,bikesLA_data['price'])
bikesLA_data['price']= np.where(bikesLA_data['price'] == '$$',1,bikesLA_data['price'])
bikesLA_data['price']= np.where(bikesLA_data['price'] == '$$$',2,bikesLA_data['price'])
bikesLA_data['price']= np.where(bikesLA_data['price'] == '$$$$',3,bikesLA_data['price'])

bikesLA_data['price'] = pd.to_numeric(bikesLA_data['price'])
bikesLA_data.info()
```

```python
#Remove columns that do not contribute to the model
df_stats = bikesLA_data.drop(columns=['latitude','longitude','uid','name'])
```

```python
#Basic information data
df_stats.describe()
```

Provide model output and an interpretation of the results. 

```python
#run full model
y = df_stats['free_bikes']
X = df_stats.drop('free_bikes', axis=1)
X = sm.add_constant(X) #adds a column of 1's so the model will contain an intercept

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())

```

> - After reviewing the data from Foursquare and Yelp against the city bike dataset; I notice that the values do not have a high correlation. For example the quantity of available bikes is not related with any other variable.
> - Even after including the price variable and review counts there is not strong dependance of the variables.
> - The model requires more information to define any conclusion. It is hard to determine if the reviews are related to people using city bikes.
